```python
import numpy as np
import matplotlib.pyplot as plt
import csv

file = ('/Users/matheus.souza/desenvolvimento/projetos/sdcnd/term2/CarND-PID-Control-Project/data/twiddle_body.txt')

plt.figure(figsize=(10,5))

with open(file) as lines:
    p = [];
    d = [];
    err = [];
    for line in lines:
        p.append(line.split()[0])
        d.append(line.split()[2])
        err.append(line.split()[8])
    
    x = np.arange(len(p))

    plt.plot(x, p, label="P Controll")
    plt.plot(x, d, label="D Controll")
    
    # Place a legend to the right of this smaller subplot.
    plt.legend(loc='best', borderaxespad=0.)
    
    plt.xlabel('iteration')
    plt.ylabel('PID controll')
    plt.title('PID')
    plt.grid(True)
    plt.savefig("pid_controll.png")

    plt.show()
```

```python
plt.figure(figsize=(10,5))

plt.plot(x, err)

plt.xlabel('iteration')
plt.ylabel('CTE')
plt.title('CTE')
plt.grid(True)
plt.savefig("cte.png")

plt.show()
```
